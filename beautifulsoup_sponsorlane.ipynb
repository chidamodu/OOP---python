```python
from urllib2 import urlopen
BASE_URL = "https://my.donationmatch.com/companies"
def cat_links(section_url):
    h = urlopen(section_url).read()
    soup = BeautifulSoup(h, "lxml")
    bob = soup.find("tbody")
    cat_links = [BASE_URL + td.a["href"] for td in bob.findAll("td")]
    return cat_links
```
